```python
import requests
from bs4 import BeautifulSoup

# get website contents
url = 'https://www.geeksforgeeks.org/response-text-python-requests/'
response = requests.get(url)

# 200 is good!!
if response.status_code == 200:
    # parse using beautiful soup
    soup = BeautifulSoup(response.text, 'html.parser')

    # get everything that is headers and paragraphs THANK YOU BEAUTIFUL SOUP ILY
    headers = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
    paragraphs = soup.find_all('p')

    # get the text from those categories
    print("Headers:")
    for header in headers:
        # strip removes extra whitespace
        print(header.get_text(strip=True))  

    # yuhhh output 
    print("\nParagraphs:")
    for paragraph in paragraphs:
        print(paragraph.get_text(strip=True))

else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```
